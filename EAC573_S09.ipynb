NUMPY - Averages and variances


````python
# LPBProgram25b.py
import numpy as np
myDATA = [11, 22, 33, 45, 77, 88, 99]
print( np.percentile(myDATA, 50))
w = [1,1,1,2,2,2,2]
print(np.median(myDATA))
print(np.average(myDATA, weights = w))
print(np.mean(myDATA))
print(np.std(myDATA))
print(np.var(myDATA))

````
